```python
# 11. Escribe una función de Python para comprobar si un número es perfecto o no.
# Según Wikipedia: En teoría de números, un número perfecto es un número entero positivo que es igual a la suma de sus divisores positivos propios,
# es decir, la suma de sus divisores positivos excluyendo el número en sí (también conocido como su alícuota suma). 
# De manera equivalente, un número perfecto es un número que es la mitad de la suma de todos sus divisores positivos (incluido él mismo).

def nperfect(num):
  sum=0
  for i in range (1, num):
    if num%i==0:
      sum=sum+i
  
  if sum==num:
      print(num, " es un numero perfecto")
  else:
    print(num, " No es un numero perfecto") 


nperfect(6)
```

```python
# 12. Escriba una función de Python que verifique si una cadena pasada es palíndromo o no. Ir al editor
# Nota: Un palíndromo es una palabra, frase o secuencia que se lee igual hacia atrás que hacia adelante.

def palindromo (cadena):
  inicio=0
  fin=len(cadena)-1
  while inicio<=fin:
    if cadena[inicio]==cadena[fin]:
      if inicio==fin:
        print(cadena, " Es palindromo :D")
        break
      fin=fin-1
      inicio=inicio+1
    else:
        print(cadena, " No es palindromo :(")
        break

palindromo('aza')
```

```python
# 14. Escriba una función de Python para verificar si una cadena es un pangrama o no. Ir al editor
# Nota: Los pangramas son palabras u oraciones que contienen cada letra del alfabeto al menos una vez.

import string, sys
#se envian el código ascci con el alfabeto y una cadena dada por el usuario a la función pangrama como argumentos
def pangrama (cadena, alphabet=string.ascii_lowercase):
    # se crea un set/ conjunto con el alfabeto
    # se convierte la cadena dada por elusuario en un set y en minusculas para hacer la comparación con abcdario
    abcdario = len(set(alphabet))
    nuevo=len(set(cadena.lower()))
    # Comparamos el némero de elementos entre ambos sets para definir si nuevo tiene almenos una vez cada letra del abcdario
    if nuevo>=abcdario:
      print("Es pangrama")
    else:
      print("No es pangrama")
 
print ( pangrama('The quick brown fox jumps over the lazy dog')) 
print ( pangrama('hola'))
```

```python
# 17. Escriba un programa en Python para crear una cadena de decoradores de funciones (negrita, cursiva, subrayado, etc.) en Python
# https://www.apsl.net/blog/2009/08/31/decoradores-en-python/#:~:text=Los%20decoradores%20forman%20parte%20de,la%20legibilidad%20y%20la%20mantenibilidad

def make_bold(fn):
    def wrapped():
        return "<b>" + fn() + "</b>"
    return wrapped

def make_italic(fn):
    def wrapped():
        return "<i>" + fn() + "</i>"
    return wrapped

def make_underline(fn):
    def wrapped():
        return "<u>" + fn() + "</u>"
    return wrapped
@make_bold
@make_italic
@make_underline
def hello():
    return "hello world"
print(hello()) ## returns "<b><i><u>hello world</u></i></b>"
```
